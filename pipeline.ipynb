```python
# Cell 7 — Interactive widget with paste support + logs
from pipeline_v5 import query_sap
from ipywidgets import widgets
from IPython.display import display
import time

text_input = widgets.Textarea(
    placeholder='Type or paste your SAP question here...',
    layout=widgets.Layout(width='90%', height='60px')
)
output_area = widgets.Output()
button = widgets.Button(
    description="Ask →",
    button_style="primary",
    layout=widgets.Layout(width='120px', height='60px')
)

def on_ask(b):
    with output_area:
        output_area.clear_output()
        q = text_input.value.strip()
        if not q:
            print("⚠️ Please enter a question")
            return

        print(f"Q: {q}")
        print("-" * 60)
        print("⏳ Generating MongoDB query (Claude)...")

        start = time.time()
        try:
            answer  = query_sap(q, verbose=True)
            elapsed = time.time() - start
            print(f"\n✅ Done in {elapsed:.1f}s")
            print("=" * 60)
            print(answer)
        except Exception as e:
            elapsed = time.time() - start
            print(f"\n❌ Error after {elapsed:.1f}s: {e}")

button.on_click(on_ask)
display(widgets.HBox([text_input, button]), output_area)
```
